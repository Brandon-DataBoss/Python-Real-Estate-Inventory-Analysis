```python
import pandas as pd
import plotly.express as px
import matplotlib as mp
import numpy as np
```

Data sourced from Realtor.com 

```python
#Import data

data = pd.read_csv("Inventory_Core_Metrics.csv")
data.head()
```

Data Cleanse 

```python
#Drop NaN row

data = data.drop(3825)
data.head()
```

```python
#Renaming date column to date_time

data.rename(columns={"month_date_yyyymm": "date_time"}, inplace=True)
data.head()
```

```python
#Check data types for all columns in the dataframe

data.dtypes.head()
```

```python
#Reformatting date column to Year - Month - Day

data["date_time"] = pd.to_datetime(data["date_time"], format="%Y%m", errors='coerce')

#Updating date column to a string

for col in data.select_dtypes(include= ['datetime64']).columns.tolist():
    data[col] = data[col].astype(str)

data.head()
```

Visualize data

```python
# Median list price (month over month)

fig_month_price = px.bar(data, 
                        x="median_listing_price", 
                        y="state", 
                        color="median_listing_price",
                        color_continuous_scale='Blugrn', 
                        animation_frame= 'date_time',
                        height = 1000,
                        template='plotly_dark', 
                        title="Median Listing Price: (month view)"
                        )
fig_month_price.update_layout(yaxis={'categoryorder':'total ascending'})
fig_month_price.show()
```

```python
#Create year column to view median list price (year over year)

conditions = [
    (data['date_time'] >= '2022 - 01 - 01'), #2022
    (data['date_time'] > '2021 - 12 - 31') & (data['date_time'] <= '2022 - 12 - 31'), #2021
    (data['date_time'] > '2020 - 12 - 31') & (data['date_time'] <= '2021 - 12 - 31'), #2020
    (data['date_time'] > '2019 - 12 - 31') & (data['date_time'] <= '2020 - 12 - 31'), #2019
    (data['date_time'] > '2018 - 12 - 31') & (data['date_time'] <= '2019 - 12 - 31'), #2018
    (data['date_time'] > '2017 - 12 - 31') & (data['date_time'] <= '2018 - 12 - 31'), #2017
    (data['date_time'] > '2016 - 12 - 31') & (data['date_time'] <= '2017 - 12 - 31')  #2016
    ]

# create a list of the values we want to assign for each condition
values = ['2022', '2021', '2020', '2019','2018', '2017', '2016']

# create a new column and use np.select to assign values to it using our lists as arguments
data['year'] = np.select(conditions, values)

# display updated DataFrame
data.head()
```

```python
#Create dataframe to visualize average median price 


yr_price = data.groupby(['year', 'state'])['median_listing_price'].mean().reset_index(name='mean_price')
yr_price.head()

```

```python
#Median list price (year over year)

fig_year_price = px.bar(yr_price, 
                        x = 'mean_price', 
                        y = 'state', 
                        color = 'mean_price',
                        color_continuous_scale='Blugrn', 
                        animation_frame= 'year',
                        height = 1000,
                        template='plotly_dark', 
                        title="Average Median Listing Price: (year view)"
                        )
fig_year_price.update_layout(yaxis={'categoryorder':'total ascending'})
fig_year_price.show()
```

```python
#Active listings by state (year over year)

fig_month_price = px.line(data, 
                        x="date_time", 
                        y="active_listing_count", 
                        color="state", 
                        animation_frame= 'state',
                        height=600,
                        template='plotly_dark', 
                        title="Active Listings"
                        )
fig_month_price.update_layout(yaxis={'categoryorder':'total ascending'},
                            yaxis_range=[0,170000]
                            )
fig_month_price.show()
```

```python
#Median dats on Market by state (year over year)

fig_month_price = px.line(data, 
                        x="date_time", 
                        y="median_days_on_market", 
                        color="state", 
                        animation_frame= 'state',
                        height=600,
                        template='plotly_dark', 
                        title="Median Days on Market"
                        )
fig_month_price.update_layout(yaxis={'categoryorder':'total ascending'},
                            yaxis_range=[0,105]
                            )
fig_month_price.show()
```

```python
#Total listings by state (year over year)

fig_month_price = px.line(data, 
                        x="date_time", y="total_listing_count", 
                        color="state", 
                        animation_frame= 'state',
                        height=600,
                        template='plotly_dark', 
                        title="Total Listings"
                        )
fig_month_price.update_layout(yaxis={'categoryorder':'total ascending'},
                            yaxis_range=[0,300000]
                            )
fig_month_price.show()
```

```python
#Active listings compared to days on market (year over year)

fig_scatter = px.scatter(data,
                        x = "active_listing_count", 
                        y="median_days_on_market",
                        trendline="ols",
                        custom_data=["date_time","state","median_listing_price"],
                        color="state",
                        size="median_listing_price",
                        animation_frame= 'year',
                        height=600, 
                        template='plotly_dark', 
                        title="Active Listings vs Days on Market"
)

#Update hover formatting 

fig_scatter.update_traces(
    hovertemplate="<br>".join([
        "active_listing_count: %{x}",
        "median_days_on_market: %{y}",
        "date_time: %{customdata[0]}",
        "state: %{customdata[1]}",
        "median_listing_price: %{customdata[2]}",
        
    ])
)
fig_scatter.show()
```
